```python
import requests
import matplotlib.pyplot as plt
from datetime import datetime, timezone
import time

latitudes = []
longitudes = []
timestamps = []

num_points = 100

url = "https://api.wheretheiss.at/v1/satellites/25544"
```

```python
for _ in range(num_points):
    response = requests.get(url)
    data = response.json()
    latitudes.append(data['latitude'])
    longitudes.append(data['longitude'])
    timestamps.append(datetime.fromtimestamp(data['timestamp'], timezone.utc))
    time.sleep(10) 
```

```python
plt.figure(figsize=(12, 6))
plt.plot(longitudes, latitudes, marker='o', markersize=2, linestyle='-', color='blue')
plt.title('Path of the International Space Station (ISS)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```
